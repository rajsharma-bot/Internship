# Q5:-Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver=webdriver.Chrome(r"chromedriver.exe")  
driver.maximize_window()
driver.get("https://www.google.com/maps/")
user_input=input("Enter city name for which coordinates is required : ")

time.sleep(5)
search=driver.find_element(By.XPATH,'//input[@autofocus="autofocus"]')
search.click()
search.send_keys(user_input)
search_button=driver.find_element(By.ID,'searchbox-searchbutton')
search_button.click()
time.sleep(5)
lat=[]
long=[]
try:
    url_string = driver.current_url
    print("URL Extracted :",url_string)
    lat_long = re.findall(r'@(.*)data',url_string)
    if len(lat_long):
        lat_long_list = lat_long[0].split(",")
        if len(lat_long_list)>=2:
            lat.append(lat_long_list[0])
            long.append(lat_long_list[1])
except Exception as e:
    print("error:",str(e))

time.sleep(2)
driver.close()

location_data=pd.DataFrame({'Enter Location':user_input,'latitude':lat,'longitude':long})
location_data
```

# 7. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver=webdriver.Chrome(r"chromedriver.exe")  
driver.maximize_window()
driver.get("https://www.digit.in/")

time.sleep(3)

laptop=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[4]/ul/li[3]")
laptop.click()
time.sleep(3)

game_laptop=driver.find_element(By.XPATH,"//a[text()='Best Gaming Laptops in India']")
game_laptop.click()

time.sleep(3)

laptop_names=[]
laptop_os=[]
laptop_display=[]
processor=[]
storage=[]
product_price = []
product_description =[]

lpName=driver.find_elements(By.XPATH,"//div[@class='TopNumbeHeading active sticky-footer']")
for i in lpName:
    lp_name=i.text.split("\n")[1]
    laptop_names.append(lp_name)
    
spec=driver.find_elements(By.XPATH,"//div[@class='Specs-Wrap']")
for i in spec:
    lp_spec=i.text.replace("OS","OS $").replace("DISPLAY","DISPLAY $").replace("PROCESSOR","PROCESSOR $").replace("\n","")
    lpSpec=lp_spec.split("$")
    laptop_os.append(lpSpec[0])
    laptop_display.append(lpSpec[1])
    processor.append(lpSpec[2])
    storage.append(lpSpec[3])
    
description = driver.find_elements(By.XPATH,"//div[@class='Section-center']")
for i in description:
    product_description.append(str(i.text).replace("\n",""))


price = driver.find_elements(By.XPATH,"//td[@class='smprice']")
for i in price:
    product_price.append(i.text.replace("\n",""))

time.sleep(5)
driver.close()

laptop_dataframe=pd.DataFrame({'Laptop Name':laptop_names,'OS':laptop_os,'Display':laptop_display,'Graphics Processor':processor,'Memory':storage,'Price':product_price,'Description':product_description})
laptop_dataframe
```

# 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video

```python
#https://www.youtube.com/watch?v=r-hZb0iVQZI
url=input("Enter the URL: ")
driver=webdriver.Chrome(r"chromedriver.exe")  
driver.maximize_window()
driver.get(url)
time.sleep(5)

comments = []
upvote = []
posted = []

for _ in range(30):
    driver.execute_script("window.scrollBy(0,100)")
    time.sleep(3)
    
    com_xpath = driver.find_elements(By.ID,"content-text")
    for i in com_xpath:
        comments.append(i.text)
        
    post = driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
    for i in post:
        posted.append(i.text)
        
        
    upvotes = driver.find_elements(By.ID,"vote-count-middle")
    for i in upvotes:
        upvote.append(i.text)

        
driver.close()      

youtube_video=pd.DataFrame({'Comments':comments,'Comments Upvote':upvote,'Comment is POSTED':posted})
youtube_video
```

```python
driver = webdriver.Chrome(r"chromedriver.exe")
driver.maximize_window()
try:

    driver.get("https://www.hostelworld.com/")

    loc = driver.find_element(
        By.XPATH,
        "/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/div/input",
    )

    loc.send_keys("london")
    search_btn = driver.find_element(
        By.XPATH,
        "/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[5]/button",
    )
    search_btn.click()

except Exception as e:

    driver.get(
        "https://www.hostelworld.com/s?q=London, England&country=England&city=London&type=city&id=3&from=2022-07-28&to=2022-07-31&guests=2&HostelNumber"
    )


time.sleep(3)

hostel_name = []
location = []
ratings = []
total_reviews = []
overall_reviews = []
private_price = []
dorm_price = []

hostel = driver.find_elements(By.XPATH, "//h2[@class='title title-6']")
try:
    for i in hostel:
        hostel_name.append(i.text)
except:
    hostel_name.append("-")


distance = driver.find_elements(By.XPATH, "//span[@class='description']")
try:
    for i in distance:
        location.append(i.text)
except:
    location.append("-")


information = driver.find_elements(By.XPATH, "//div[@class='info']/div[2]")
try:
    for i in information:
        ratings.append(i.text.split("\n")[0])
except:
    ratings.append("-")

reviews = driver.find_elements(By.XPATH, "//div[@class='reviews']")
try:
    for i in reviews:
        total_reviews.append(i.text)
except:
    total_reviews.append("-")


review = driver.find_elements(By.XPATH, "//div[@class='keyword']/span")
try:
    for i in review:
        overall_reviews.append(i.text)
except:
    overall_reviews.append("-")


private = driver.find_elements(By.XPATH, "//a[@class='prices']/div[1]")
try:
    for i in private:
        private_price.append(i.text.replace("\n", ""))
except:
    private_price.append("-")


dorm = driver.find_elements(By.XPATH, "//a[@class='prices']/div[2]")
for i in dorm:
    dorm_price.append(i.text.replace("\n", ""))


property_description = []
prop_url = []
url = driver.find_elements(By.XPATH, "//h2[@class='title title-6']/a")
for i in url:
    prop_url.append(i.get_attribute("href"))

for i in prop_url:
    driver.get(i)
    time.sleep(5)
    try:
        description = driver.find_element(
            By.XPATH,
            "/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[2]/div/div/div[2]",
        )
        property_description.append(description.text)
    except NoSuchElementException:
        property_description.append("-")

time.sleep(5)
driver.close()

available_hostel = pd.DataFrame(
    {
        "Hostel Name": hostel_name,
        "Distance from City Center": location,
        "Ratings": ratings,
        "Total Review": total_reviews,
        "Overall Review": overall_reviews[0:30],
        "Private Price": private_price[0:30],
        "Dorm Price": dorm_price,
        "Description": property_description,
        "URL": prop_url,
    }
)

available_hostel
```
